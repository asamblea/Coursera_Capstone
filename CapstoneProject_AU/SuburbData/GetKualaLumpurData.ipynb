```python
url = "https://en.wikipedia.org/wiki/Kuala_Lumpur#Local_government"
soup = BeautifulSoup(requests.get(url).text, 'html.parser')

results = soup.find_all('div', attrs={'class':'div-col columns column-width'})
# results
suburbs = []
for result in results:
    for i in result.find_all('a', href=True):
        suburbs.append(i.get('title'))

# suburbs
```

```python
df_suburbs=pd.DataFrame(suburbs, columns=['Address'])
df_suburbs['Address'] = df_suburbs['Address'].str.replace('(, ).*','')

df_suburbs['Address'] += ', Kuala Lumpur Malaysia'
df_suburbs[['Suburb','Region']] = df_suburbs.Address.str.split(", ", expand=True,)

df_suburbs.head()
```
