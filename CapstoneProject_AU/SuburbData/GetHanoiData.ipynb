```python
url = "https://en.wikipedia.org/wiki/Hanoi"
soup = BeautifulSoup(requests.get(url).text, 'html.parser')

results = soup.find_all('tbody')[5].find_all('td')
results
suburbs = []
for result in results:
    # i = result.find_all('a', href=True)
    for i in result.find_all('a', href=True):
        suburbs.append(i.get('title'))
        
# get the 12 urban districts only
suburbs = suburbs[0:12]
suburbs
```

```python
df_suburbs=pd.DataFrame(suburbs, columns=['Address'])
df_suburbs['Address'] = df_suburbs['Address'].str.replace(' District','')

df_suburbs['Address'] += ', Hanoi Vietnam'
df_suburbs[['Suburb','Region']] = df_suburbs.Address.str.split(", ", expand=True,)

df_suburbs
```
