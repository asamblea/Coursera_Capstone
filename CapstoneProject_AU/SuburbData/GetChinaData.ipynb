```python
url_beijing = "https://en.wikipedia.org/wiki/Neighborhoods_in_Beijing"
soup_beijing = BeautifulSoup(requests.get(url_beijing).text, 'html.parser')

results = soup_beijing.find_all('div', attrs={'class':'div-col columns column-width'})[0].find_all('a', href=True)
# results
```

```python
suburbs = []
for result in results:
    suburbs.append(result.get('title') + ', Beijing China')
# suburbs
```

```python
url_shanghai = "https://en.wikipedia.org/wiki/Category:Neighbourhoods_of_Shanghai"
soup_shanghai = BeautifulSoup(requests.get(url_shanghai).text, 'html.parser')

results2 = soup_shanghai.find_all('li')
results2

sh = []
for result in results2:
    for i in result.find_all('a', href=True):
        if i.get('title') is not None:
            sh.append(i.get('title')+', Shanghai China')

for n in range(0,19):
    suburbs.append(sh[n])
suburbs
```

```python

df_suburbs=pd.DataFrame(suburbs, columns=['Address'])
df_suburbs['Address'] = df_suburbs['Address'].str.replace(', Beijing,',',')
df_suburbs['Address'] = df_suburbs['Address'].str.replace(', Shanghai,',',')

df_suburbs['Address'] = df_suburbs['Address'].str.replace(r" \(.*\)","")
# df_suburbs['Address'] += ', Beijing China'
df_suburbs[['Suburb','Region']] = df_suburbs.Address.str.split(", ", expand=True,)

df_suburbs.drop_duplicates(keep='first',inplace=True)
# df_suburbs=df_suburbs.replace(regex=r'Pudong', value='Pudong New District')
df_suburbs
```
