```python
url = "https://en.wikipedia.org/wiki/Tokyo#Sister_cities,_sister_states,_and_friendship_agreements"
soup = BeautifulSoup(requests.get(url).text, 'html.parser')

results = soup.find_all('tbody')[3].find_all('td')
results
suburbs = []
for result in results:
    # i = result.find_all('a', href=True)
    for i in result.find_all('a', href=True):
        suburbs.append(i.get('title'))
        
suburbs
```

```python
df_suburbs=pd.DataFrame(suburbs, columns=['Address'])
df_suburbs['Address'] = df_suburbs['Address'].str.replace(', Tokyo','')

df_suburbs['Address'] += ', Tokyo Japan'
df_suburbs[['Suburb','Region']] = df_suburbs.Address.str.split(", ", expand=True,)
df_suburbs = df_suburbs.drop([1],axis=0)
df_suburbs.head()
```
