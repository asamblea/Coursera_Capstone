```python
url = 'https://geographic.org/streetview/new_zealand/canterbury_west_coast/christchurch/index.html'
soup = BeautifulSoup(requests.get(url).text, 'html.parser')
results = soup.find_all('li')

results

suburbs = []
for result in results:
    suburbs.append(result.text.strip() + ', Christchurch New Zealand')
# suburbs
```

```python

url2 = 'https://en.wikipedia.org/wiki/List_of_suburbs_of_Auckland'
soup2 = BeautifulSoup(requests.get(url2).text, 'html.parser')
results2 = soup2.find_all('div', attrs={'class':'div-col columns column-width'})[0].find_all('a', href=True)
results2
for result in results2:
    suburbs.append(result.get('title') + ', Auckland City New Zealand')

# suburbs
```

```python
df_suburbs=pd.DataFrame(suburbs, columns=['Address'])
df_suburbs['Address'] = df_suburbs['Address'].str.replace(', New Zealand','')
df_suburbs['Address'] = df_suburbs['Address'].str.replace(', Auckland,',',')
df_suburbs['Address'] = df_suburbs['Address'].str.replace(r" \(.*\)","")
df_suburbs[['Suburb','Region']] = df_suburbs.Address.str.split(", ", expand=True,)

df_suburbs.head()
```
