```python
url = "https://en.m.wikipedia.org/wiki/London_boroughs"
soup = BeautifulSoup(requests.get(url).text, 'html.parser')

results = soup.find_all('td')[-2:]
results
suburbs = []
for result in results:
    for i in result.find_all('a', href=True):
        suburbs.append(i.get('title'))
        
suburbs = suburbs[1:]
```

```python
df_suburbs=pd.DataFrame(suburbs, columns=['Address'])
df_suburbs['Address']=df_suburbs['Address'].str.split('Borough of').str[-1]
df_suburbs['Address'] += ', London UK'
df_suburbs[['Suburb','Region']] = df_suburbs.Address.str.split(", ", expand=True)
df_suburbs
```
