```python
url = "https://en.m.wikipedia.org/wiki/Neighbourhoods_of_Delhi"
soup = BeautifulSoup(requests.get(url).text, 'html.parser')

# results = soup.find_all('div',{'class':'mw-parser-output'})
results = soup.find_all('ul')[6:15]

results

suburbs = []
for result in results:
    for i in result.find_all('a', href=True):
        t = i.get('title')
        if t != None:
            suburbs.append(t + ', New Delhi India')

# suburbs
```

```python
url2 = 'https://en.wikipedia.org/wiki/List_of_neighbourhoods_in_Mumbai#South_Mumbai'
soup2 = BeautifulSoup(requests.get(url2).text, 'html.parser')
results2 = soup2.find_all('ul')[28]

results2

for result in results2.find_all('a', href=True):
    suburbs.append(result.get('title')+ ', Mumbai City India')
    
# suburbs
```

```python

df_suburbs=pd.DataFrame(suburbs, columns=['Address'])
df_suburbs['Address'] = df_suburbs['Address'].str.replace(r" \(.*\)",'')
df_suburbs['Address'] = df_suburbs['Address'].str.replace(', Delhi,',',')
df_suburbs['Address'] = df_suburbs['Address'].str.replace(', New Delhi,',',')
df_suburbs['Address'] = df_suburbs['Address'].str.replace(', Mumbai,',',')

# df_suburbs['Address'] += ', Delhi India'
df_suburbs[['Suburb','Region']] = df_suburbs['Address'].str.split(', ', expand=True)
df_suburbs.shape
# df_suburbs.query('Address.str.contains(",")')
```
