```python
from bs4 import BeautifulSoup
import requests
from urllib.request import urlopen

url_brisbane = "https://en.wikipedia.org/wiki/City_of_Brisbane"
soup_brisbane = BeautifulSoup(requests.get(url_brisbane).text, 'html.parser')

results = soup_brisbane.find_all('div', attrs={'class':'div-col columns column-width'})
# results
suburbs_brisbane = []
for result in results:
    for i in result.find_all('a', href=True):
        suburbs_brisbane.append(i.get('title'))

# suburbs_brisbane
```

```python
df_suburbs_brisbane=pd.DataFrame(suburbs_brisbane, columns=['address'])
# df_suburbs_brisbane.loc[(~df_suburbs_brisbane['address'].str.contains('New South Wales')),'address'] +=', New South Wales'
df_suburbs_brisbane['address'] = df_suburbs_brisbane['address'].str.replace(', Queensland','')
df_suburbs_brisbane=df_suburbs_brisbane.replace(regex=r'central business district', value='CBD')
df_suburbs_brisbane['address'] += ', Brisbane'
df_suburbs_brisbane[['Suburb','Region']] = df_suburbs_brisbane.address.str.split(", ", expand=True,)
# df_suburbs_brisbane.drop(df_suburbs_brisbane[df_suburbs_brisbane['Suburb'].str.contains('Council|Randwick|Woollahra')].index,inplace=True)

df_suburbs_brisbane.head()
```

```python
address = 'City of Brisbane, Australia'

geolocator = Nominatim(user_agent="brisbane_explorer")
location = geolocator.geocode(address)
brisbane_latitude = location.latitude
brisbane_longitude = location.longitude
print('The geograpical coordinate of brisbane are {}, {}.'.format(brisbane_latitude, brisbane_longitude))

# create map of brisbane using latitude and longitude values
map_brisbane = folium.Map(location=[brisbane_latitude, brisbane_longitude], zoom_start=10)

# add markers to map
for lat, lng, borough, neighborhood in zip(df_suburbs_brisbane['Latitude'], df_suburbs_brisbane['Longitude'], df_suburbs_brisbane['Region'], df_suburbs_brisbane['Suburb']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_brisbane)

map_brisbane
```
